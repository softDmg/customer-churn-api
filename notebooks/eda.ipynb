```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/maidawa/Desktop/Damags/Projects/customer_churn/data/raw/telco_churn.csv")

df.head(5)

print(df[0:5])
```

```python
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
df["Churn"].value_counts(normalize=True).plot(kind="bar", color=["skyblue", "orange"])
plt.title("Churn Rate Distribution")
plt.xticks([0,1], ['No Churn', 'Churn'], rotation=0)
plt.xlabel("Churn")
plt.ylabel("Proportion")
plt.show()
```

```python
sns.histplot(data=df, x="tenure", hue="Churn", multiple="stack")
plt.title("Tenure Distribution by Churn")
plt.show()
```

```python
sns.boxplot(data=df, x="Churn", y="MonthlyCharges")
plt.title("Monthly Charges by Churn")
plt.show()
```

```python
pd.crosstab(df["Contract"], df["Churn"], normalize='index').plot(kind='bar', stacked=True)
plt.title("Churn by Contract Type")
plt.ylabel("Proportion")
plt.show()
```
